```python
import pyspark
```

```python
import re
```

```python
from operator import add
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName('parta').setMaster('local')
sc = SparkContext(conf = conf)
```

```python
dataset1 = sc.textFile('Agile Processes  in Software Engineering  and Extreme Programming.txt') 
dataset2 = sc.textFile('Scrum Handbook.txt')
```

```python
no_of_lines1 = len(dataset1.collect())
no_of_lines1
```

```python
no_of_lines2 = len(dataset2.collect())
no_of_lines2
```

```python
def remove(content):
    pattern = r"([^a-zA-Z\s]+)"
    return re.sub(pattern, "", content)
```

```python
alphabet_dataset1 = dataset1.map(remove)
alphabet_dataset2 = dataset2.map(remove)
```

```python
alphabet_dataset1.take(10)
```

```python
alphabet_dataset2.take(10)
```

```python
def remove_trailing_heading(content):
    pattern = r"(\s+$|^\s+)"
    return re.sub(pattern, "", content)
```

```python
remove_trailing_heading_dataset1 = alphabet_dataset1.map(remove_trailing_heading)
remove_trailing_heading_dataset2 = alphabet_dataset2.map(remove_trailing_heading)
```

```python
remove_trailing_heading_dataset1.take(10)
```

```python
remove_trailing_heading_dataset2.take(10)
```

```python
def lower(lines):
    lines = lines.lower()
    return lines
```

```python
lower_dataset1 = remove_trailing_heading_dataset1.map(lower)
lower_dataset2 = remove_trailing_heading_dataset2.map(lower)
```

```python
lower_dataset1.collect()
```

```python
lower_dataset2.collect()
```

```python
split_dataset1 = lower_dataset1.flatMap(lambda line: line.split(" "))
split_dataset2 = lower_dataset2.flatMap(lambda line: line.split(" "))
```

```python
split_dataset1.take(10)
```

```python
split_dataset2.take(10)
```

```python
pair_dataset1 = split_dataset1.map(lambda key: (key, 1))
pair_dataset2 = split_dataset2.map(lambda key: (key, 1))
```

```python
pair_dataset1.take(10)
```

```python
pair_dataset2.take(10)
```

```python
sort_dataset1 = pair_dataset1.reduceByKey(add).map(lambda element: (element[1], element[0])).sortByKey(ascending = False)
sort_dataset2 = pair_dataset2.reduceByKey(add).map(lambda element: (element[1], element[0])).sortByKey(ascending = False)
```

```python
sort_dataset1.take(20)
```

```python
sort_dataset2.take(20)
```

```python
stop_word = set(stopwords.words('english'))
```

```python
remove_stopword_dataset1 = pair_dataset1.filter(lambda element: element[0] not in stop_word)
remove_stopword_dataset2 = pair_dataset2.filter(lambda element: element[0] not in stop_word)
```

```python
remove_stopword_dataset1.take(10)
```

```python
remove_stopword_dataset2.take(10)
```

```python
unique_dataset1 = remove_stopword_dataset1.reduceByKey(add).filter(lambda element: element[1] == 1)
unique_dataset2 = remove_stopword_dataset2.reduceByKey(add).filter(lambda element: element[1] == 1)
```

```python
unique_dataset1.take(10)
```

```python
unique_dataset2.take(10)
```

```python
unique_dataset1.count()
```

```python
unique_dataset2.count()
```

```python
average_occurrence_dataset1 = unique_dataset1.count()/(split_dataset1.count() - split_dataset1.collect().count(""))
average_occurrence_dataset2 = unique_dataset2.count()/(split_dataset2.count() - split_dataset2.collect().count(""))
```
